```python
import os

from openclean.data.load import stream

ds = stream(os.path.join('data', '43nn-pn8j.tsv.gz'))
```

```python
df = ds.to_df()
```

Pamdas iterrows() is very slow (see for example)

https://stackoverflow.com/questions/16476924/how-to-iterate-over-rows-in-a-dataframe-in-pandas

```python
import time

start = time.perf_counter()

for _, row in df.iterrows():
    pass

end = time.perf_counter()

print('{} s'.format(end - start))
```

```python
import time

start = time.perf_counter()

for i, rowid in enumerate(df.index):
    df.iat[i, 1]

end = time.perf_counter()

print('{} s'.format(end - start))
```

```python
import time

from openclean.function.eval.base import Col
from openclean.operator.transform.filter import filter

start = time.perf_counter()

bronx = filter(df, Col('BORO') == 'Bronx')

end = time.perf_counter()

print('{} s'.format(end - start))
```
